```python
import pandas as pd
import os

#os.chdir("C:\\Users\\gokul\\Documents\\learning\\practise problem\\airbnb\\Reviews and listings")
os.chdir("C:\\Users\\ggovinda\\Documents\\Untitled Folder\\airbnb-listings")

listings = pd.read_csv("listings.csv")

listings.info()
```

```python
drop_listings=['listing_url', 'experiences_offered', 'space', 'neighborhood_overview', 'notes', 
               'transit', 'access', 'interaction', 'house_rules', 'thumbnail_url', 'medium_url', 
               'picture_url', 'xl_picture_url', 'host_url','experiences_offered', 'neighbourhood_group_cleansed', 
               'has_availability', 'license', 'jurisdiction_names', 'host_thumbnail_url', 'host_picture_url', 
               'neighborhood_overview', 'neighbourhood', 'cleaning_fee', 
               'space','notes', 'transit', 'access', 'interaction', 'house_rules', 'host_about', 
               'first_review', 'security_deposit', 
              'last_review', 'weekly_price', 'monthly_price', 'square_feet']
listings = listings.drop(drop_listings, axis=1)
listings.info()
```

```python
listings.isnull().sum()
```

```python
import numpy as np


listings['reviews_per_month'] = listings['reviews_per_month'].replace(np.NaN, listings['reviews_per_month'].mean())
listings['review_scores_rating'] = listings['review_scores_rating'].replace(np.NaN, listings['review_scores_rating'].mean())
listings['review_scores_accuracy'] = listings['review_scores_accuracy'].replace(np.NaN, listings['review_scores_accuracy'].mean())
listings['review_scores_cleanliness'] = listings['review_scores_cleanliness'].replace(np.NaN, listings['review_scores_cleanliness'].mean())
listings['review_scores_checkin'] = listings['review_scores_checkin'].replace(np.NaN, listings['review_scores_checkin'].mean())
listings['review_scores_communication'] = listings['review_scores_communication'].replace(np.NaN, listings['review_scores_communication'].mean())
listings['review_scores_location'] = listings['review_scores_location'].replace(np.NaN, listings['review_scores_location'].mean())
listings['review_scores_value'] = listings['review_scores_value'].replace(np.NaN, listings['review_scores_value'].mean())
```

```python
listings.fillna(0, inplace=True)
listings.isnull().sum()

#listings[['host_id','host_location']]
         
new_listings = listings[['host_id','smart_location','room_type','price','latitude','longitude','description','review_scores_rating']]
print(new_listings.info())
print("room_type:")
print(new_listings['room_type'].unique())
print("smart_location:")
print(new_listings['smart_location'].unique())

print(new_listings.head(5))

#replace the 


new_listings['room_type'] = new_listings['room_type'].replace(['Entire home/apt'], 'home')
new_listings['room_type'] = new_listings['room_type'].replace(['Private room'], 'private')
new_listings['room_type'] = new_listings['room_type'].replace(['Shared room'], 'shared')

print("after datamanipulating for room type")
print(new_listings['room_type'].unique())
```
